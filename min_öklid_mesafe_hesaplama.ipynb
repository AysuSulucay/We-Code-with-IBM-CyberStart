# Minimum klid Mesafesinin Hesaplanmas



- klid mesafesi, klid uzayndaki iki nokta arasndaki "sradan" dz izgi mesafesidir. Bu forml ile, dzlemde veya  boyutlu uzayda iki nokta arasndaki mesafeyi bulabilirsiniz.

**Greviniz:**

- Python'da fonksiyonlar ve dngler kavramlarn kullanarak, aadaki ilemleri gerekletiren bir program yazmanz gerekmektedir:

### Noktalarn Tanmlanmas:

- points adnda bir liste oluturun. Bu liste, 2D uzaydaki noktalar temsil eden demetler (tuple) iermelidir. rnein, (x, y) noktas bir demet (x, y) olarak temsil edilecektir.

```python
points = [(8, 2), (4, 16), (3, 7), (12,50)]
```

### klid Mesafesi in Bir Fonksiyon Yazma:

- euclideanDistance adnda bir fonksiyon tanmlayn. Bu fonksiyon, iki demet (her biri bir noktay temsil eder) almal ve bu iki nokta arasndaki klid mesafesini dndrmelidir.

**Forml:**


```python
import math

def euclideanDistance(point1, point2):
    return math.sqrt((point2[0] - point1[0]) ** 2 + (point2[1] - point1[1]) ** 2)
```

### Mesafelerin Hesaplanmas:

- Bir dng kullanarak, points listesindeki her nokta ifti arasndaki klid mesafesini hesaplayn. Bu mesafeleri distances adnda baka bir listede saklayn.

```python
distances = []
for i in range(len(points)):
    for j in range(i + 1, len(points)):
        dist = euclideanDistance(points[i], points[j])
        distances.append((dist, points[i], points[j]))
```

```python
# Mesafeleri ve hangi noktalar arasnda hesaplandklarn yazdrmak
for distance in distances:
    print(f"Mesafe: {distance[0]:.2f} - Noktalar: {distance[1]} ve {distance[2]}")
```

### Minimum Mesafenin Bulunmas:

- distances listesinden minimum mesafeyi bulun ve yazdrn.

```python
# Minimum mesafeyi bul ve hangi noktalar arasnda olduunu yazdr
min_distance = min(distances, key=lambda x: x[0])  # Mesafeye gre minimumu bulur
print(f"Minimum klid Mesafesi: {min_distance[0]:.2f} - {min_distance[1]} , {min_distance[2]} noktalarna ait olan mesafe.")
```
